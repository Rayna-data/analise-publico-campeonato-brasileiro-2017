```python
df['data'] = pd.to_datetime(df['data'], dayfirst=True)
```

```python
df.sort_values('publico', ascending=False).head(10)
```

```python
maior_publico = df['publico'].max()
menor_publico = df['publico'].min()

print(f"Maior público: {maior_publico}")
print(f"Menor público: {menor_publico}")
```

```python
df.loc[df['publico'].idxmin()]
```

```python
top_10 = df.sort_values('publico', ascending=False).head(10)
```
